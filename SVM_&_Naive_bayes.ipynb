#Q1 What is a Support Vector Machine (SVM)?


A **Support Vector Machine (SVM)** is a supervised machine learning algorithm used for **classification, regression, and outlier detection**. It is particularly effective for high-dimensional data and is widely used in fields like text classification, image recognition, and bioinformatics.

### **How SVM Works**
1. **Hyperplane and Decision Boundary**  
   - SVM finds the best **hyperplane** that separates data points into different classes.
   - This hyperplane maximizes the **margin** (the distance between the nearest data points from each class, known as **support vectors**).

2. **Linear vs. Non-Linear SVM**  
   - **Linear SVM**: If data is linearly separable, SVM finds a straight-line (or hyperplane in higher dimensions) to classify the data.
   - **Non-Linear SVM**: If data is not linearly separable, SVM uses **kernel tricks** (such as polynomial, radial basis function (RBF), or sigmoid kernels) to transform data into higher dimensions where it becomes separable.

3. **Mathematical Formulation**  
   - The SVM algorithm solves an **optimization problem** to minimize classification errors while maximizing the margin.
   - It uses **Lagrange multipliers** and **quadratic programming** to achieve this.

4. **Regularization Parameter (C)**  
   - Controls the trade-off between maximizing the margin and minimizing classification errors.
   - A **high C** results in a smaller margin with fewer misclassifications.
   - A **low C** allows more margin but may misclassify some points.

### **Advantages of SVM**
- Effective for **high-dimensional data**.
- Works well for **small datasets** with complex patterns.
- **Robust to overfitting** in high-dimensional space.
- Can handle **non-linear data** with the right kernel.

### **Disadvantages of SVM**
- Computationally expensive for **large datasets**.
- Choosing the right **kernel** and hyperparameters can be tricky.
- Doesnâ€™t perform well when **data is highly imbalanced**.








#Q2 What is the difference between Hard Margin and Soft Margin SVM?




The main difference between **Hard Margin SVM** and **Soft Margin SVM** lies in how they handle data that is not perfectly separable.

### 1. **Hard Margin SVM**:
   - **Strictly Separable Data**: Hard margin SVM assumes that the data is **linearly separable**, meaning there exists a perfect hyperplane that can completely separate the data points of one class from another without any errors or overlaps.

   - **No Misclassification Allowed**: In this approach, no data points can be on the wrong side of the hyperplane. The algorithm attempts to find a hyperplane that maximizes the margin while ensuring that all data points are correctly classified.
   - **Limitations**:
     - This method is highly sensitive to noise and outliers in the dataset, as even a single misclassified point would make it impossible to find a valid hyperplane.

     - Not suitable for real-world scenarios where data often isn't perfectly separable.

### 2. **Soft Margin SVM**:
   - **Allowing Misclassification**: Soft margin SVM is a more flexible version that **allows some misclassification** of data points. The algorithm introduces a penalty for misclassifications, which helps balance between maximizing the margin and minimizing the misclassification errors.

   - **Penalty Parameter (C)**: The **C parameter** controls the trade-off between maximizing the margin and minimizing the classification error. A **large C** will make the model behave more like a hard margin SVM (i.e., it will prioritize minimizing misclassifications). A **small C** allows more flexibility, leading to a larger margin but potentially more misclassified points.
   - **Suitable for Noisy or Overlapping Data**: Soft margin SVM is more practical for real-world datasets where classes are often not perfectly separable. It can handle situations where there are outliers or where the data is noisy.







#Q3 What is the mathematical intuition behind SVM?


The **mathematical intuition** behind **Support Vector Machines (SVM)** centers around the idea of finding the optimal hyperplane that maximizes the margin between two classes of data points. Let's break it down step by step:

### 1. **Hyperplane in Feature Space**:
   - A **hyperplane** is a decision boundary that separates different classes in the feature space. For a 2D problem, the hyperplane is simply a **line**, and in a 3D problem, it's a **plane**. In general, for higher-dimensional spaces, it remains a **hyperplane**.
   - Mathematically, a hyperplane can be described by the equation:
     $$
     \mathbf{w} \cdot \mathbf{x} + b = 0
     $$
     where:
     - $\mathbf{w}$ is a vector perpendicular to the hyperplane (also called the **normal vector**).
     - $b$ is the **bias** term that determines the position of the hyperplane.
     - $\mathbf{x}$ is any point on the feature space.

### 2. **Maximizing the Margin**:
   - The goal of SVM is to find the hyperplane that **maximizes the margin** between the two classes. The **margin** is defined as the distance between the hyperplane and the closest data points from either class, which are called **support vectors**.
   - If the data points are perfectly separable, the support vectors lie on the two parallel hyperplanes that are the closest to the separating hyperplane:
     $$
     \mathbf{w} \cdot \mathbf{x}_i + b = 1 \quad \text{for points in class +1}
     $$
     $$
     \mathbf{w} \cdot \mathbf{x}_i + b = -1 \quad \text{for points in class -1}
     $$
     Here, $\mathbf{x}_i$ represents the support vectors for each class.

### 3. **Objective: Maximize the Margin**:
   The **margin** is the distance between these two hyperplanes:
   $$
   \text{Margin} = \frac{2}{\|\mathbf{w}\|}
   $$
   The distance between the two parallel hyperplanes is $ \frac{2}{\|\mathbf{w}\|} $. Therefore, to maximize the margin, we need to **minimize** $\|\mathbf{w}\|$, which is equivalent to **maximizing the margin**.

   Thus, the objective function to minimize is:
   $$
   \text{Minimize} \quad \frac{1}{2} \|\mathbf{w}\|^2
   $$
   The factor of $ \frac{1}{2} $ simplifies the math in the optimization process, as we only need to minimize $ \|\mathbf{w}\|^2 $ instead of $ \|\mathbf{w}\| $.

### 4. **Optimization Problem (Hard Margin)**:
   - The SVM needs to find the optimal hyperplane that separates the data such that no points are on the wrong side of the margin. For perfectly separable data, the optimization problem becomes:
     $$
     \text{Minimize} \quad \frac{1}{2} \|\mathbf{w}\|^2
     $$
     subject to the constraint that for each training point $\mathbf{x}_i$, the correct classification is achieved:
     $$
     y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1
     $$
     where $y_i \in \{-1, +1\}$ is the label of the data point. This constraint ensures that all points are correctly classified and lie on the correct side of the margin.

### 5. **Soft Margin (for Non-linearly Separable Data)**:
   In real-world scenarios, the data might not be perfectly separable, so we introduce **slack variables** ($\xi_i$) that allow some data points to lie on the wrong side of the margin. The idea is to allow a small margin of error while still maximizing the margin as much as possible.

   The optimization problem becomes:
   $$
   \text{Minimize} \quad \frac{1}{2} \|\mathbf{w}\|^2 + C \sum_{i=1}^n \xi_i
   $$
   where $C$ is a regularization parameter that controls the trade-off between maximizing the margin and minimizing the classification error. The term $C \sum_{i=1}^n \xi_i$ penalizes misclassifications, and the slack variables $\xi_i$ represent how far each point is allowed to be on the wrong side of the margin.

   The constraints for the soft margin SVM are now:
   $$
   y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 - \xi_i
   $$
   and
   $$
   \xi_i \geq 0
   $$
   This allows some points to violate the margin, but the model will try to minimize the violations (penalized by the $\xi_i$) while still maximizing the margin.

### 6. **Kernel Trick (Non-linear SVM)**:
   In cases where the data is not linearly separable in the original feature space, SVM can use the **kernel trick**. The idea is to map the data into a higher-dimensional space where a linear hyperplane can separate the data points.

   The kernel function $K(\mathbf{x}_i, \mathbf{x}_j)$ computes the inner product of the data points in the transformed feature space, without explicitly computing the transformation. Common kernels include:
   - **Linear kernel**: $K(\mathbf{x}_i, \mathbf{x}_j) = \mathbf{x}_i \cdot \mathbf{x}_j$
   - **Polynomial kernel**: $K(\mathbf{x}_i, \mathbf{x}_j) = (\mathbf{x}_i \cdot \mathbf{x}_j + c)^d$
   - **Radial Basis Function (RBF) kernel**: $K(\mathbf{x}_i, \mathbf{x}_j) = \exp(-\gamma \|\mathbf{x}_i - \mathbf{x}_j\|^2)$

   Using a kernel function, the optimization problem still looks similar, but instead of working directly with the feature vectors, we work with their transformed versions in the higher-dimensional space.




#Q4 What is the role of Lagrange Multipliers in SVM?


In **Support Vector Machines (SVM)**, Lagrange multipliers play a critical role in formulating the optimization problem that defines the optimal hyperplane for classification tasks.

The main goal of SVM is to find a hyperplane that separates two classes of data with the maximum margin, i.e., the distance between the hyperplane and the closest points from both classes (known as support vectors). To achieve this, the SVM optimization problem can be formulated as a **constrained optimization problem**, which is where Lagrange multipliers come into play.

### Role of Lagrange Multipliers in SVM:

1. **Formulating the Optimization Problem**:
   The SVM optimization problem is typically written as:
   
   $$
   \min_{\mathbf{w}, b} \frac{1}{2} \|\mathbf{w}\|^2
   $$

   Subject to the constraints:

   $$
   y_i (\mathbf{w}^T \mathbf{x}_i + b) \geq 1, \quad \text{for all } i = 1, 2, \dots, N
   $$

   Where:
   - $ \mathbf{w} $ is the weight vector (normal to the hyperplane),
   - $ b $ is the bias term (defines the offset),
   - $ \mathbf{x}_i $ is a data point, and
   - $ y_i $ is the label of the data point, where $ y_i \in \{-1, 1\} $.

2. **Introducing Lagrange Multipliers**:
   To handle the inequality constraints $ y_i (\mathbf{w}^T \mathbf{x}_i + b) \geq 1 $, Lagrange multipliers $ \lambda_i $ are introduced for each constraint. This allows the transformation of the constrained optimization problem into an unconstrained one.

   The Lagrangian function is constructed as:

   $$
   \mathcal{L}(\mathbf{w}, b, \lambda) = \frac{1}{2} \|\mathbf{w}\|^2 - \sum_{i=1}^N \lambda_i \left[ y_i (\mathbf{w}^T \mathbf{x}_i + b) - 1 \right]
   $$

   Where $ \lambda_i \geq 0 $ are the Lagrange multipliers.

3. **KKT Conditions**:
   The **Karush-Kuhn-Tucker (KKT)** conditions are used to solve this problem. The KKT conditions include:
   - **Stationarity**: The gradient of the Lagrangian with respect to $ \mathbf{w} $ and $ b $ must be zero.
   - **Primal feasibility**: The constraints must be satisfied.
   - **Dual feasibility**: The Lagrange multipliers must be non-negative.
   - **Complementary slackness**: For each constraint, $ \lambda_i $ can only be non-zero if the corresponding constraint is active, i.e., $ y_i (\mathbf{w}^T \mathbf{x}_i + b) = 1 $.

4. **Dual Problem**:
   By applying these conditions, the SVM optimization problem can be rewritten in a dual form, which is easier to solve, especially when using kernel methods. The dual formulation maximizes the following function:

   $$
   \max_{\lambda} \sum_{i=1}^N \lambda_i - \frac{1}{2} \sum_{i,j=1}^N \lambda_i \lambda_j y_i y_j \mathbf{x}_i^T \mathbf{x}_j
   $$

   The dual problem involves the Lagrange multipliers $ \lambda_i $, and it helps identify the support vectors (points where $ \lambda_i > 0 $) that define the margin.

Support vectors in a **Support Vector Machine (SVM)** are the data points that are crucial for defining the optimal decision boundary (also called the hyperplane) that separates the different classes in the dataset.

Hereâ€™s a breakdown:

- **Decision boundary (Hyperplane)**: In SVM, the goal is to find the hyperplane that best separates the classes in the data. For a 2D dataset, this is a line, and for higher dimensions, itâ€™s a plane or higher-dimensional hyperplane.

- **Support vectors**: These are the data points that are closest to the decision boundary. They are the "support" or foundation for the decision boundary. In other words, they lie on the edges of the margin (the region between the two classes), and they help determine the position and orientation of the hyperplane.

- **Importance of support vectors**: The key property of SVM is that it tries to maximize the margin, the distance between the closest data points of each class to the decision boundary. The support vectors directly influence this margin, and they are the only points used to determine the hyperplane. This makes SVM quite efficient because it doesnâ€™t need all the data points, only the support vectors.

### Visualizing Support Vectors:
Imagine a 2D space where you have two classes of points. The SVM algorithm will find a line that separates these points, and the support vectors are the points closest to this line. These points are critical because if you were to remove them, the position of the decision boundary could change.

#Q7 What is a Support Vector Regressor (SVR)?


A **Support Vector Regressor (SVR)** is a machine learning algorithm used for regression tasks, where the goal is to predict a continuous output variable based on input features. It is based on the **Support Vector Machine (SVM)** algorithm, which is commonly used for classification tasks, but SVR is adapted for predicting numerical values.

Hereâ€™s a breakdown of the key components:

### 1. **Support Vectors**:
   - The concept of "support vectors" comes from the classification version of SVM. In SVR, support vectors are the data points that are closest to the regression line (or hyperplane in higher dimensions) and play a crucial role in defining the model. These points are the ones that influence the decision boundary, i.e., the line or curve that predicts the output.

### 2. **Epsilon-Insensitive Tube**:
   - SVR uses a concept called an **epsilon-insensitive tube** (often just called the epsilon margin), where a certain margin of tolerance is allowed around the predicted values. Within this margin, errors are not penalized. Only errors outside this margin (above epsilon) contribute to the loss function.
   - The parameter `epsilon (Îµ)` controls this margin of tolerance. The larger the epsilon, the fewer points will be penalized for errors.

### 3. **Objective**:
   - The objective of SVR is to find a function that deviates from the actual target values (in the training data) by a margin no greater than epsilon, while at the same time being as flat as possible. This "flatness" helps avoid overfitting, ensuring good generalization to unseen data.
   
### 4. **Regularization Parameter (C)**:
   - **C** is a regularization parameter that determines the trade-off between a low error and a low model complexity. A small value of C allows more error but leads to a simpler model, while a larger value of C forces the model to fit the training data more precisely, potentially increasing the risk of overfitting.

### 5. **Kernel Trick**:
   - Just like SVM for classification, SVR can also employ the **kernel trick**. By using kernels like the **linear**, **polynomial**, or **RBF (Radial Basis Function)** kernels, SVR can model non-linear relationships between the input features and the target variable, making it versatile for a wide range of data types.

### 6. **Loss Function**:
   - SVR uses a loss function that is different from traditional regression models. It uses a **Îµ-insensitive loss function**, meaning that it only penalizes errors larger than Îµ. The goal is to minimize the number of points outside the Îµ-tube and ensure the model is as simple as possible.

### 7. **Training Process**:
   - The training of the SVR involves solving an optimization problem to minimize a cost function that balances fitting the data with the complexity of the model (the number of support vectors used).

### Advantages of SVR:
- **Robust to overfitting**, especially in high-dimensional spaces.
- Can model **non-linear relationships** through kernel functions.
- **Effective in high-dimensional spaces** and when there is a clear margin of separation.

### Disadvantages of SVR:
- **Computationally expensive** for large datasets, especially when using complex kernels.
- **Choice of parameters** (like C, Îµ, and kernel) can significantly impact the model's performance and may require tuning.

### Use Cases:
SVR is typically used in applications where the target variable is continuous, and you need to model complex, non-linear relationships. Examples include:
- Predicting stock prices.
- Estimating house prices.
- Forecasting demand for products or services.
















#Q8 What is the Kernel Trick in SVM?


The **Kernel Trick** in **Support Vector Machines (SVM)** is a mathematical technique that allows SVM to efficiently perform non-linear classification by implicitly mapping the original data into a higher-dimensional feature space, without actually computing the coordinates in that higher-dimensional space. This is useful for handling complex, non-linearly separable data.

Hereâ€™s a breakdown of the concept:

### 1. **Linear SVMs and Their Limitation**:
SVM is a classifier that works by finding the hyperplane that best separates data into different classes. For **linearly separable** data, this is straightforward: SVM finds a hyperplane that maximizes the margin (distance) between data points of different classes. However, when data is not linearly separable (i.e., it cannot be separated by a straight line or hyperplane), SVM struggles.

### 2. **Non-linear Data**:
For non-linearly separable data, the **Kernel Trick** allows SVM to create a decision boundary that works in higher dimensions, where the data might become linearly separable.

For example, consider data in a 2D space that cannot be separated by a straight line. By applying a transformation (using a **kernel function**) to map the data to a higher-dimensional space, the transformed data might become linearly separable, allowing SVM to find an appropriate hyperplane.

### 3. **How the Kernel Trick Works**:
Instead of explicitly computing the mapping of data to a higher-dimensional space, the kernel function computes the **inner product** (dot product) of the data points in this higher-dimensional space directly. This avoids the computational cost of explicitly mapping the data points, which is why itâ€™s called the "trick."

### 4. **Common Kernel Functions**:
Some popular kernel functions include:

- **Linear Kernel**: No transformation, equivalent to standard linear SVM.
  
  $$
  K(x, y) = x^T y
  $$

- **Polynomial Kernel**: Maps data to a higher-dimensional space using polynomials.
  
  $$
  K(x, y) = (x^T y + c)^d
  $$

  where $c$ and $d$ are constants.

- **Radial Basis Function (RBF) or Gaussian Kernel**: Maps data into an infinite-dimensional space. It's widely used for SVM.
  
  $$
  K(x, y) = \exp\left(-\frac{\|x - y\|^2}{2\sigma^2}\right)
  $$

- **Sigmoid Kernel**: Uses a sigmoid function, similar to a neural network activation.
  
  $$
  K(x, y) = \tanh(\alpha x^T y + c)
  $$

  where $\alpha$ and $c$ are constants.

### 5. **Benefits of the Kernel Trick**:
- **Efficient computation**: We don't need to explicitly calculate the higher-dimensional feature space, which could be computationally expensive.
- **Non-linear classification**: Enables SVM to handle complex, non-linear classification tasks by implicitly mapping data into higher dimensions.
- **Flexibility**: By choosing different kernel functions, we can adjust the SVM's decision boundaries to suit the problem at hand.

### 6. **Example**:
Imagine you have two classes of data that are circularly arranged (non-linearly separable). By using an RBF kernel, you can transform the data into a higher-dimensional space where it becomes separable by a linear hyperplane, even though it was not separable in the original 2D space.


#Q9 Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.



Linear, Polynomial, and Radial Basis Function (RBF) kernels are all different ways to compute the similarity between data points in a Support Vector Machine (SVM) or other kernel-based algorithms. Each kernel has its strengths and weaknesses depending on the nature of the data. Here's a comparison:

### 1. **Linear Kernel**

**Formula**:
$$
K(x, x') = x \cdot x'
$$

- **Description**: The linear kernel is the simplest of the three. It computes the dot product of the input vectors. If the data is already linearly separable (can be separated by a straight line or hyperplane), the linear kernel performs well.
  
- **When to use**:
  - When data is linearly separable.
  - Computationally efficient for high-dimensional data.
  
- **Advantages**:
  - Fast and simple.
  - Works well when the data is linearly separable or close to it.
  - No hyperparameters to tune other than the regularization parameter ($C$).

- **Disadvantages**:
  - Cannot handle complex relationships in the data (if the data is not linearly separable).
  - Not suitable for highly complex, nonlinear patterns.

### 2. **Polynomial Kernel**

**Formula**:
$$
K(x, x') = (x \cdot x' + c)^d
$$
Where $c$ is a constant (typically non-negative), and $d$ is the degree of the polynomial.

- **Description**: The polynomial kernel computes the similarity between two data points by taking the dot product, adding a constant, and then raising it to a power. This allows the model to capture more complex patterns than the linear kernel by introducing nonlinearity.

- **When to use**:
  - When data is not linearly separable but the complexity of the decision boundary can be modeled by a polynomial function.
  - Suitable for problems where interactions between features are expected.
  
- **Advantages**:
  - Can model more complex decision boundaries than a linear kernel.
  - Flexible (tuning the degree $d$ and constant $c$ allows for a variety of curves and shapes).

- **Disadvantages**:
  - Computationally more expensive than the linear kernel, especially with higher degrees.
  - Overfitting can occur with higher-degree polynomials if not tuned properly.

### 3. **Radial Basis Function (RBF) Kernel**

**Formula**:
$$
K(x, x') = \exp\left(-\frac{\|x - x'\|^2}{2\sigma^2}\right)
$$
Where $\sigma$ is a free parameter (often denoted as $\gamma$) that controls the width of the kernel.

- **Description**: The RBF kernel computes the similarity between two points based on their Euclidean distance, transformed by a Gaussian function. This is highly flexible and can capture very complex decision boundaries, making it a popular choice in many real-world SVM applications.

- **When to use**:
  - When the decision boundary is highly nonlinear.

  - Common choice when you don't have prior knowledge about the data distribution and need flexibility.
  
- **Advantages**:
  - Can handle complex, nonlinear relationships.

  - Very flexible in terms of the decision boundary shape.
  - Works well even if the decision boundary is not clearly defined by any linear or polynomial relationship.

- **Disadvantages**:
  - Can be computationally expensive, especially with large datasets.

  - Requires careful tuning of the $\gamma$ parameter. If $\gamma$ is too large, the model may overfit; if it's too small, it may underfit.
  - Sensitive to the choice of $\gamma$ and may need cross-validation to find the optimal value.



### When to Choose Each Kernel:
- **Linear kernel**: If your data is already linearly separable or if you have high-dimensional sparse data like text classification (e.g., TF-IDF).

- **Polynomial kernel**: If you believe the decision boundary could be captured by polynomial relationships and you want to capture feature interactions.

- **RBF kernel**: If you need a flexible and general-purpose kernel that can capture complex, nonlinear relationships without much domain knowledge.

In practice, **RBF** is often a go-to kernel because of its versatility, but it's important to tune the $\gamma$ parameter carefully to avoid overfitting or underfitting.











#Q10 What is the effect of the C parameter in SVM?



The **C parameter** in Support Vector Machines (SVM) plays a critical role in controlling the trade-off between maximizing the margin (distance between the separating hyperplane and the nearest data points) and minimizing classification errors (misclassifications).

Here's how it affects the model:

1. **Large C Value (Low Regularization):**
   - **Penalty for misclassification is high**: When C is large, the SVM model will try to classify all data points correctly, even if it means creating a smaller margin. This leads to less tolerance for misclassifications.
   - **Overfitting risk**: The model might overfit the training data because it focuses heavily on correctly classifying every point, potentially at the expense of generalization.

2. **Small C Value (High Regularization):**
   - **Allowing some misclassifications**: When C is small, the model will tolerate some misclassifications in order to achieve a larger margin between the classes.
   - **Underfitting risk**: The model might become too simple (underfit), as it prioritizes a larger margin over getting all the classifications correct. It might not capture the complexities of the data.









#Q11 What is the role of the Gamma parameter in RBF Kernel SVM?


The **gamma** parameter in the **RBF (Radial Basis Function) kernel** of a Support Vector Machine (SVM) controls the influence of a single training example and determines the shape of the decision boundary in a non-linear SVM.

### Here's how gamma affects the SVM model:

1. **High Gamma (Large Value):**
   - **Closer influence of data points**: A high value of gamma means that each training point's influence is more localized and falls off very quickly as you move away from the point.

   - **Sharper decision boundary**: This leads to a more complex model with a decision boundary that closely fits the training data, as each point has a strong influence over the classification.
   - **Risk of overfitting**: The model may overfit, capturing noise or outliers in the data, because it tries to classify each point with high precision.

2. **Low Gamma (Small Value):**
   - **Wider influence of data points**: A low gamma means that each data point has a broader influence, and the decision boundary becomes smoother.

   - **Smoother decision boundary**: The SVM model becomes less sensitive to individual data points and generalizes better to unseen data.
   - **Risk of underfitting**: If gamma is too low, the model might not capture the underlying patterns in the data, leading to underfitting.

### In summary:
- **High Gamma** â†’ More localized influence â†’ Complex decision boundary â†’ Risk of overfitting.
- **Low Gamma** â†’ More global influence â†’ Smoother decision boundary â†’ Risk of underfitting.

The ideal gamma value depends on the dataset. A typical approach is to use cross-validation to find the optimal value of gamma for the best generalization performance.







#Q12 What is the NaÃ¯ve Bayes classifier, and why is it called "NaÃ¯ve"?




The **NaÃ¯ve Bayes classifier** is a probabilistic machine learning algorithm used for classification tasks, such as spam detection, sentiment analysis, or medical diagnosis. It is based on **Bayes' Theorem**, which describes the probability of an event based on prior knowledge of conditions related to the event.

### How NaÃ¯ve Bayes Works:
1. **Bayes' Theorem**: The classifier calculates the probability of a class $C$ given a set of features $X = (x_1, x_2, ..., x_n)$ using the formula:
   $$
   P(C | X) = \frac{P(X | C) \cdot P(C)}{P(X)}
   $$
   - $P(C | X)$: The posterior probability of class $C$ given the features $X$.

   - $P(X | C)$: The likelihood of the features $X$ given class $C$.
   - $P(C)$: The prior probability of class $C$.
   - $P(X)$: The probability of the features $X$ (acts as a normalizing constant).

2. **NaÃ¯ve Assumption**: The algorithm assumes that all features $x_1, x_2, ..., x_n$ are **conditionally independent** given the class $C$. This means:
   $$
   P(X | C) = P(x_1 | C) \cdot P(x_2 | C) \cdot ... \cdot P(x_n | C)
   $$
   This simplifies the calculation of $P(X | C)$, as it only requires estimating the probability of each feature individually.

3. **Classification**: For a given set of features, the classifier calculates the posterior probability for each class and assigns the instance to the class with the highest probability.

### Why is it Called "NaÃ¯ve"?
The term "naÃ¯ve" refers to the algorithm's simplifying assumption that all features are **conditionally independent** given the class. In reality, features in a dataset are often correlated or dependent on each other. By ignoring these dependencies, the algorithm makes a "naÃ¯ve" assumption, which simplifies the computation but may not always reflect the true relationships in the data.

Despite this simplification, NaÃ¯ve Bayes often performs surprisingly well in practice, especially in text classification tasks, because:
- It is computationally efficient.
- It works well with high-dimensional data (e.g., text data with many words).
- It requires relatively little training data to estimate parameters.

- **Example: Spam Detection**
  - Consider classifying emails as spam or not spam based on words like "Free", "Win", "Offer".
  - NaÃ¯ve Bayes assumes each word contributes independently to the probability of spam.
  - In reality, words often appear together (e.g., "Win a free iPhone"), but NaÃ¯ve Bayes still performs well.

#Q13 What is Bayesâ€™ Theorem?


**Bayes' Theorem** is a fundamental principle in probability theory that describes how to update the probability of a hypothesis based on new evidence. It provides a way to calculate the probability of an event (or hypothesis) given some observed data, incorporating prior knowledge and new information.

The formula for Bayes' Theorem is:

$$
P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)}
$$

Where:
- $P(A | B)$ is the **posterior probability**: the probability of hypothesis $A$ given the data $B$.
- $P(B | A)$ is the **likelihood**: the probability of observing the data $B$ given that hypothesis $A$ is true.
- $P(A)$ is the **prior probability**: the initial probability of hypothesis $A$ before observing the data.
- $P(B)$ is the **marginal likelihood** or **evidence**: the total probability of observing the data $B$ under all possible hypotheses.

### Intuition behind Bayes' Theorem:
- **Prior Probability ($P(A)$)**: This represents what we know or believe about the probability of $A$ before any evidence is taken into account. For example, if weâ€™re trying to predict whether someone has a disease, the prior might be based on the general prevalence of the disease in the population.
  
- **Likelihood ($P(B | A)$)**: This tells us how likely the observed data $B$ is, assuming that $A$ is true. Continuing the disease example, it could represent how likely certain symptoms are given that the person actually has the disease.

- **Posterior Probability ($P(A | B)$)**: This is the updated probability of the hypothesis $A$ (like whether the person has the disease) after considering the new evidence $B$ (like the symptoms observed).

- **Evidence ($P(B)$)**: This is the probability of observing the data $B$ under all possible hypotheses. It acts as a normalizing factor, ensuring that the posterior probabilities across all hypotheses sum to 1.

### Example:
Suppose you're a doctor trying to determine whether a patient has a particular disease based on a positive test result. Let's say:
- The prior probability of having the disease ($P(A)$) is 1% (this is based on the general prevalence of the disease).
- The likelihood ($P(B | A)$) is the probability of a positive test result given that the person has the disease, which is 90%.
- The probability of a positive test result ($P(B)$) across all people is 10% (it includes both people who have the disease and those who don't).

You want to know: **Given a positive test result, what is the probability the patient has the disease?**

Using Bayes' Theorem:

$$
P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)} = \frac{0.9 \times 0.01}{0.1} = 0.09
$$

This means that even with a positive test result, there's only a 9% chance that the patient actually has the disease. This result is surprising, but it illustrates how important the prior probability and the overall prevalence of the condition are when interpreting diagnostic test results.

### Applications:
Bayes' Theorem is widely used in various fields, including:
- **Medical Diagnosis**: Calculating the likelihood of a disease given test results.
- **Machine Learning**: NaÃ¯ve Bayes classifiers use Bayes' Theorem for classification tasks.
- **Spam Filtering**: Determining whether an email is spam based on features like the presence of certain words.
- **Decision Making**: Updating beliefs or decisions based on new evidence.













#Q14 Explain the differences between Gaussian NaÃ¯ve Bayes, Multinomial NaÃ¯ve Bayes, and Bernoulli NaÃ¯ve Bayes.



The differences between **Gaussian NaÃ¯ve Bayes**, **Multinomial NaÃ¯ve Bayes**, and **Bernoulli NaÃ¯ve Bayes** primarily lie in how they model the feature distributions in the data. Let's break each of them down:

### 1. **Gaussian NaÃ¯ve Bayes (GNB)**
- **Distribution Assumption**: Assumes that the features (attributes) follow a **Gaussian (normal) distribution**. This means that for each feature, the likelihood is modeled as a Gaussian (bell curve) distribution.
- **Use Case**: This classifier is used when the features are continuous and have a distribution that is approximately normal (bell-shaped).
- **How It Works**: It calculates the probability of a feature belonging to a particular class by assuming that the feature's values for each class are distributed normally (with a mean and variance specific to each class).
  
  **Example**: If you have features like height, weight, or temperature, which are continuous and could roughly follow a Gaussian distribution, GNB is a good choice.

### 2. **Multinomial NaÃ¯ve Bayes (MNB)**
- **Distribution Assumption**: Assumes that the features are **counts** or represent frequencies of occurrences. Specifically, it assumes that the features follow a **Multinomial distribution** (which is appropriate for categorical data that represents counts or frequencies).
- **Use Case**: This model is typically used in text classification tasks where the features are word counts (e.g., in spam filtering or document classification), or when features are discrete counts of events.
- **How It Works**: It models the likelihood of the feature counts for each class by using the multinomial distribution, which is suitable for counting occurrences (like how many times a word occurs in a document).

  **Example**: In text classification, each feature could represent the count of a particular word in a document. MNB would model the probability of seeing that count given a class.

### 3. **Bernoulli NaÃ¯ve Bayes (BNB)**
- **Distribution Assumption**: Assumes that the features are **binary** (i.e., they take only two possible values, often 0 or 1). The features are modeled using the **Bernoulli distribution**, which is suitable for binary data.
- **Use Case**: This is useful when the features represent the presence or absence of a certain attribute (like whether a word is present in a document, or whether a certain event occurred).
- **How It Works**: It calculates the probability of a binary feature being present or absent in a given class, using the Bernoulli distribution.

  **Example**: In a spam email classification task, a feature could represent whether a specific word is present (1) or absent (0) in the email, and Bernoulli NaÃ¯ve Bayes would model this binary occurrence.

#Q16 What are the key assumptions made by NaÃ¯ve Bayes?



NaÃ¯ve Bayes is called "naÃ¯ve" because it makes **simplifying assumptions** about the data that may not always hold in reality. These assumptions make the model computationally efficient while still performing well in many cases.



### **1. Feature Independence Assumption**
- **NaÃ¯ve Bayes assumes that all features are independent of each other**, given the class label.  
- Mathematically, for a given class ($C$), the probability of a data point with features $( X_1, X_2, ..., X_n$) is:

  \$
  P(X_1, X_2, ..., X_n | C) = P(X_1 | C) \cdot P(X_2 | C) \cdot ... \cdot P(X_n | C)
  \$

- **Real-world example:**  
  - In spam detection, words like **"free"** and **"win"** often appear together in spam emails.
  - NaÃ¯ve Bayes treats them **independently**, even though they are correlated.



### **2. Conditional Independence Assumption**
- Each feature contributes **independently** to the classification **after knowing the class label**.  
- This means that **given the class (Spam or Not Spam), the presence of one word does not affect the probability of another word appearing**.

- **Example (Text Classification):**  
  - If we classify emails as **Spam ($S$)** or **Not Spam ($NS$)**, NaÃ¯ve Bayes assumes:

    \$
    P("win", "prize" | S) = P("win" | S) \times P("prize" | S)
    \$

  - Even if "win" and "prize" often appear together, NaÃ¯ve Bayes **ignores this dependency**.



### **3. Equal Importance of Features**
- NaÃ¯ve Bayes **assumes all features contribute equally** to determining the class.  
- In reality, some features may be more important than others.

- **Example:**  
  - In email classification:
    - The word **"urgent"** may be more indicative of spam than **"meeting"**.
    - NaÃ¯ve Bayes **treats both words as equally important**, which may not always be ideal.



### **4. Data Follows a Specific Distribution**  
- Depending on the type of NaÃ¯ve Bayes classifier used, it assumes the data follows a certain **probability distribution**:
  - **Gaussian NaÃ¯ve Bayes** assumes features are **normally distributed** (bell-shaped curve).

  - **Multinomial NaÃ¯ve Bayes** assumes **word counts/frequencies** (used for text classification).
  - **Bernoulli NaÃ¯ve Bayes** assumes **binary features** (word presence/absence in documents).

- **Example:**  
  - **Gaussian NaÃ¯ve Bayes:** If predicting heights of people, NaÃ¯ve Bayes assumes heights follow a **normal distribution**.
  - **Multinomial NaÃ¯ve Bayes:** Used in spam filtering, assuming word occurrences follow a **count-based probability distribution**.



### **Limitations of These Assumptions**  
- **Feature Independence is Unrealistic** â€“ In real-world data, features are often correlated.  
- **Equal Importance of Features is Not Always True** â€“ Some features matter more than others.  
- **Incorrect Distribution Assumptions Can Affect Accuracy** â€“ If data is not normally distributed, Gaussian NaÃ¯ve Bayes may perform poorly.

However, despite these limitations, **NaÃ¯ve Bayes works surprisingly well in many practical applications** like spam filtering and text classification.

#Q19 Compare SVM and NaÃ¯ve Bayes for classification tasks.

### **1. Core Principle:**

  - Based on **statistical learning theory**, it tries to find the **optimal hyperplane** that maximizes the margin between different classes in the feature space.

### **2. Assumptions:**

  - Assumes that features are **conditionally independent** given the class, which often does not hold in real-world datasets. However, it still works well in many practical applications, especially with text classification.

  - Does **not** make strong assumptions about feature independence or distribution.

#Q20 How does Laplace Smoothing help in NaÃ¯ve Bayes?




Laplace smoothing (also known as **additive smoothing**) is a technique used to handle the problem of **zero probability** in NaÃ¯ve Bayes classifiers, especially when dealing with **discrete data** (like in text classification or other count-based feature spaces). It helps to ensure that no probability is ever exactly zero, which can be problematic for classification.



### **Problem: Zero Probability**

In NaÃ¯ve Bayes, the probability of a class given a set of features is computed using **Bayes' Theorem**:

\$
P(C|X) = \frac{P(X|C) P(C)}{P(X)}
\$

Where:
- $P(C|X)$ is the posterior probability of the class $C$ given the features $X$.
- $P(X|C)$ is the likelihood of the features $X$ given the class $C$.
- $P(C)$ is the prior probability of the class $C$.
- $P(X)$ is the probability of the features $X$ (also called the evidence).

In a typical NaÃ¯ve Bayes setup, we assume that the features $x_1, x_2, ..., x_n$ are independent, so:

\$
P(X|C) = P(x_1|C) \times P(x_2|C) \times ... \times P(x_n|C)
\$

Each of these probabilities $P(x_i|C)$ is typically estimated from the training data. However, if any feature $x_i$ does not appear in the training data for a particular class $C$, then the probability $P(x_i|C)$ will be zero. This causes the entire product $P(X|C)$ to become zero, leading to a **zero probability for the class**.

This can happen in problems like **text classification**: If a particular word (feature) does not appear in the training data for a class (e.g., the word "free" might not appear in the "ham" emails during training), the probability for that class will be zero for new documents containing that word, which is undesirable.

### **How Laplace Smoothing Works**

Laplace smoothing addresses this issue by adding a small constant (usually 1) to all feature counts in the data, ensuring that no feature has a zero probability. This way, even if a feature doesnâ€™t appear in the training set for a particular class, it will still have a non-zero probability.

The smoothed estimate for the probability of a feature $x_i$ given class $C$ is given by:

\$
P(x_i|C) = \frac{\text{count}(x_i, C) + 1}{\sum_{x} \text{count}(x, C) + |V|}
\$

Where:
- **count($x_i$, $C$)** is the number of times feature $x_i$ appears in class $C$.
- **$|V|$** is the size of the vocabulary (i.e., the number of unique features or words in the dataset).
- The **+1** in the numerator represents the Laplace smoothing term.

#### Key Points:
- The **+1** in the numerator ensures that no feature has a zero probability.
- The denominator is the total number of feature occurrences in class $C$, plus **$|V|$** to account for all possible features (words). This normalizes the probability to ensure it still sums to 1 across all features for each class.

### **Example:**

- Letâ€™s say we have a simple text classification problem where we are classifying documents as either **spam** or **ham** (not spam). In the training data, we might encounter the following frequencies for two words (e.g., "free" and "offer"):






| Word   | Spam Count | Ham Count |
|--------|------------|-----------|
| Free   | 3          | 0         |
| Offer  | 0          | 2         |




Without smoothing, if we try to calculate the probability of the word "free" given the class **ham** ($P(\text{"free"}|\text{ham})$), we would get:

\$
P(\text{"free"}|\text{ham}) = \frac{\text{count}(\text{"free"}, \text{ham})}{\text{total words in ham}} = \frac{0}{2} = 0
\$

This would result in a zero probability for the word "free" in the **ham** class, which is problematic when we classify new documents that contain the word "free".

**With Laplace smoothing**, we adjust the probability calculation as follows:

\$
P(\text{"free"}|\text{ham}) = \frac{\text{count}(\text{"free"}, \text{ham}) + 1}{\text{total words in ham} + |V|}
= \frac{0 + 1}{2 + 4} = \frac{1}{6}
\$



Here, **$|V| = 4$** because there are 4 unique words in our vocabulary: "free", "offer", "ham", and "spam".

This means the word "free" will still contribute a non-zero probability, even though it was never seen in the **ham** class in the training data.

```python
#24 Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary.


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_classification
x, y = make_classification(n_samples=500, n_clusters_per_class=2, n_features=2,n_classes=2 , n_redundant=0)


# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=1)


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

#Model training
from sklearn.svm import SVC
svc = SVC(kernel='poly')
svc.fit(x_train,y_train)

y_pred = svc.predict(x_test)


# Create a grid of points for decision boundary
x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),  # Increased resolution
                     np.linspace(y_min, y_max, 300))

# Predict for each grid point
Z = svc.predict(scaler.transform(np.c_[xx.ravel(), yy.ravel()]))
Z = Z.reshape(xx.shape)

# Plot decision boundary and data points
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.5, cmap=plt.cm.coolwarm)
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, cmap=plt.cm.coolwarm, edgecolors='k', s=20, label="Train Data")
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, cmap=plt.cm.coolwarm, marker="*", s=80, edgecolors='k', label="Test Data")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVM Decision Boundary (Polynomial Kernel)")
plt.legend()
plt.show()
```

```python
#27 Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Target'] = data.target

# Select only two features for visualization
selected_features = ['mean radius', 'mean texture']
x = df[selected_features]
y = df['Target']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

# Standardize the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Define different C values
C_values = [0.001, 0.01, 0.1, 1, 10, 100]

# Create a figure with subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 12))  # 2 rows, 3 columns

for i, c in enumerate(C_values):
    model = SVC(C=c, kernel='linear')  # Use linear kernel for visualization
    model.fit(x_train, y_train)

    # Make predictions
    y_pred = model.predict(x_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy for C={c}: {accuracy:.4f}")

    # Create a mesh grid
    h = .02  # Step size in the mesh
    x_min, x_max = x_train[:, 0].min() - 1, x_train[:, 0].max() + 1
    y_min, y_max = x_train[:, 1].min() - 1, x_train[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Predict over the entire grid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Get subplot location
    ax = axes[i // 3, i % 3]

    # Plot decision boundary
    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    ax.scatter(x_train[:, 0], x_train[:, 1], c=y_train, edgecolors='k', cmap=plt.cm.coolwarm)
    ax.set_title(f"C={c} (Acc: {accuracy:.4f})")
    ax.set_xlabel(selected_features[0])
    ax.set_ylabel(selected_features[1])
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)

# Adjust layout and show the figure
plt.tight_layout()
plt.show()
```

```python
#35 Write a Python program to perform feature selection before training a NaÃ¯ve Bayes classifier and compare results.


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris

# Load dataset
data = load_iris()
X = data.data  # Feature matrix
y = data.target  # Target variable

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# 1. Train NaÃ¯ve Bayes classifier without feature selection
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_before = nb.predict(X_test)
accuracy_before = accuracy_score(y_test, y_pred_before)
print(f"Accuracy before feature selection: {accuracy_before * 100:.2f}%")

# 2. Perform feature selection (SelectKBest using chi-squared test)
selector = SelectKBest(chi2, k=2)  # Selecting top 2 features
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# 3. Train NaÃ¯ve Bayes classifier after feature selection
nb_selected = GaussianNB()
nb_selected.fit(X_train_selected, y_train)
y_pred_after = nb_selected.predict(X_test_selected)
accuracy_after = accuracy_score(y_test, y_pred_after)
print(f"Accuracy after feature selection: {accuracy_after * 100:.2f}%")

# Compare results
if accuracy_after > accuracy_before:
    print("Feature selection improved the model accuracy.")
else:
    print("Feature selection did not improve the model accuracy.")
```

```python
#36 Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO) strategies on the Wine dataset and compare
#   their accuracy.


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_wine
data = load_wine()

df = pd.DataFrame(data.data, columns= data.feature_names)
df['Target'] = data.target

x = df.iloc[:,:-1]
y = df.iloc[:,-1]

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=1)

# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

# --- Model Training SVM Classifier using One-vs-Rest (OvR) ---
clf = SVC(decision_function_shape='ovr')
clf.fit(x_train_scaled, y_train)
y_ovr = clf.predict(x_test_scaled)
accuracy_after = accuracy_score(y_test, y_ovr)
print(f"Accuracy (One-vs-rest (OvR)): {accuracy_after * 100:.2f}%")

# --- SVM Classifier using One-vs-One (OvO) ---
clf_ovo = SVC(decision_function_shape='ovo')
clf_ovo.fit(x_train_scaled, y_train)
y_pred_ovo = clf_ovo.predict(x_test_scaled)
accuracy_ovo = accuracy_score(y_test, y_pred_ovo)
print(f"Accuracy (One-vs-One (OvO)): {accuracy_ovo * 100:.2f}%")

```

```python
#38 Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_wine
data = load_wine()

df = pd.DataFrame(data.data, columns= data.feature_names)
df['Target'] = data.target

x = df.iloc[:,:-1]
y = df.iloc[:,-1]


from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline

# Create a pipeline (scaling + SVM)
pipeline = make_pipeline(
    StandardScaler(),
    SVC(kernel='linear', random_state=42)
)

# Perform cross-validation (Stratified K-Fold by default with cv=5)
cv_scores = cross_val_score(pipeline, x, y, cv=5, scoring='accuracy')

# Compute average accuracy
average_accuracy = np.mean(cv_scores)
print(f"Average Accuracy using Stratified K-Fold: {average_accuracy * 100:.2f}%")
```

```python
#39 Write a Python program to train a NaÃ¯ve Bayes classifier using different prior probabilities and compare performance.


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_classification
x , y = make_classification(n_samples=1000, n_features=10, n_clusters_per_class=2, n_informative=2, n_classes=2, random_state=1)

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=1)

# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import BernoulliNB

# Define different prior probabilities
priors = [
    None,                  # Default (learned from data)
    [0.7, 0.3],            # Higher prior for class 0
    [0.3, 0.7]             # Higher prior for class 1
]


# Train and evaluate with different priors
for prior in priors:
    clf = BernoulliNB(alpha=1.0, binarize=0.0, class_prior=prior)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)

    if prior is None:
        print(f"Accuracy with default priors: {accuracy * 100:.2f}%")
    else:
        print(f"Accuracy with priors {prior}: {accuracy * 100:.2f}%")
```

```python
#40 Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_classification
x , y = make_classification(n_samples=1000, n_features=10, n_clusters_per_class=2, n_informative=2, n_classes=2, random_state=1)

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=1)

# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)


from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score


# --- Step 1: Train an SVM Classifier without feature selection ---
clf_full = SVC(kernel='linear', random_state=42)
clf_full.fit(x_train_scaled, y_train)
y_pred_full = clf_full.predict(x_test_scaled)
accuracy_full = accuracy_score(y_test, y_pred_full)
print(f"Accuracy without RFE: {accuracy_full * 100:.2f}%")

# --- Step 2: Apply RFE (Recursive Feature Elimination) ---
# Use an SVM classifier as the estimator for RFE
selector = RFE(estimator=SVC(kernel='linear', random_state=42), n_features_to_select=5)
selector = selector.fit(x_train_scaled, y_train)

# Get the selected features
x_train_rfe = selector.transform(x_train_scaled)
x_test_rfe = selector.transform(x_test_scaled)

# Train the SVM classifier on the selected features
clf_rfe = SVC(kernel='linear', random_state=42)
clf_rfe.fit(x_train_rfe, y_train)
y_pred_rfe = clf_rfe.predict(x_test_rfe)
accuracy_rfe = accuracy_score(y_test, y_pred_rfe)

print(f"Accuracy with RFE (selected features): {accuracy_rfe * 100:.2f}%")
```

```python
#46 Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_classification
x, y = make_classification(n_samples=1000, n_features=10, n_clusters_per_class=2, n_informative=2, n_classes=2, random_state=1)

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=1)

# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)


# Train an SVM classifier (SVC)
clf = SVC(probability=True)  # Set probability=True to use predict_proba for ROC and PR curves
clf.fit(x_train, y_train)

y_prob = clf.predict_proba(x_test)[:, 1]  # Probabilities for class 1


from sklearn.metrics import precision_recall_curve, auc
# Compute Precision-Recall curve
precision, recall, _ = precision_recall_curve(y_test, y_prob)

# Compute AUC for Precision-Recall curve
pr_auc = auc(recall, precision)

# Plot Precision-Recall curve
plt.figure(figsize=(8,6))
plt.plot(recall, precision, color='blue', label=f'Precision-Recall curve (AUC = {pr_auc:.2f})')
plt.fill_between(recall, precision, color='lightblue', alpha=0.5)
plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```
